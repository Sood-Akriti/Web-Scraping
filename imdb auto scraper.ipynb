```python
from autoscraper import AutoScraper
```

```python
imdb_url="https://www.imdb.com/search/title/?release_date=2017&sort=num_votes,desc&page=1"

wanted_list=["Logan","(2017)","77","8.1","726,124"]
```

```python
scraper=AutoScraper()
result=scraper.build(imdb_url,wanted_list)
```

```python
print(result)
```

```python
scraper.get_result_similar(imdb_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_ajm5':'Title','rule_ooen':'year','rule_trjs':'metascore','rule_1ysy':'ratings','rule_380k':'votes'})
scraper.keep_rules(['rule_ajm5','rule_ooen','rule_trjs','rule_1ysy','rule_380k'])
scraper.save('imdb-search')
```

```python
results=scraper.get_result_similar('https://www.imdb.com/search/title/?release_date=2018&sort=num_votes,desc&page=1',group_by_alias=True)
```
